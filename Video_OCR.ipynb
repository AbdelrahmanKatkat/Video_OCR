<summary>1. Introduction</summary>

**1. Introduction to Paddle OCR**

Paddle OCR is an Optical Character Recognition (OCR) toolkit developed by Baidu's PaddlePaddle deep learning platform. It aims to provide developers with easy-to-use tools for text detection, recognition, and other related tasks. Paddle OCR leverages deep learning models to accurately identify and extract text from images, making it suitable for various applications like document scanning, image translation, and text extraction from images.

**2. Text Detection**

Text detection is the process of identifying regions in an image where text is present. Paddle OCR employs deep learning techniques, including convolutional neural networks (CNNs), to detect text regions accurately. These networks analyze the input image and output bounding boxes that enclose the detected text regions.

Paddle OCR provides several pre-trained models for text detection, including models based on popular architectures like EAST (Efficient and Accurate Scene Text Detector) and DB (Deep Bonding Box).

**3. Text Recognition**

Once text regions are detected, the next step is text recognition, which involves recognizing and transcribing the text from the detected regions. Paddle OCR utilizes deep learning models, such as recurrent neural networks (RNNs) or convolutional recurrent neural networks (CRNNs), for text recognition tasks.

The recognition models are trained on large datasets of annotated text images to learn to accurately transcribe text from various fonts, styles, and languages.

**4. Supported Languages and Scripts**

Paddle OCR supports multiple languages and scripts, making it versatile for use in different regions and applications. The toolkit includes models trained on datasets containing text in languages like English, Chinese, Japanese, Korean, and others. Additionally, Paddle OCR can handle various scripts, including Latin, Cyrillic, Arabic, and East Asian characters.

**5. Customization and Fine-tuning**

One of the strengths of Paddle OCR is its flexibility and extensibility. Developers can fine-tune existing pre-trained models or train custom models on domain-specific datasets to improve performance on specialized tasks or unique text styles.

Paddle OCR provides tools and documentation to facilitate model customization and training, allowing developers to adapt the OCR system to their specific requirements.

**6. Integration and Deployment**

Paddle OCR is designed to be easily integrated into existing software applications and workflows. It provides APIs and software development kits (SDKs) for popular programming languages like Python, making it accessible to a wide range of developers.

Additionally, Paddle OCR can be deployed in various environments, including on-premises servers, cloud platforms, and edge devices, enabling real-time text recognition and processing in diverse applications.

**7. Performance and Scalability**

Paddle OCR is optimized for performance and scalability, leveraging the parallel computing capabilities of modern hardware architectures, including GPUs and multi-core CPUs. The toolkit is capable of processing large volumes of images efficiently, making it suitable for batch processing tasks and high-throughput applications.

Furthermore, Paddle OCR continuously benefits from advancements in deep learning research and optimizations in the PaddlePaddle platform, ensuring state-of-the-art performance and accuracy in text detection and recognition tasks.

```python
import re
import sys
n=0

for temp in cap_list:

  # Get the frame rate of the video
  fps = int(temp.get(cv2.CAP_PROP_FPS))

  # List and COunters
  fps_counter=0
  Vol=[]
  Cur=[]
  Pwr=[]
  Ene=[]
  Cap=[]
  Ex_Temp=[]
  Timing=[]

  while temp.isOpened():
      ret, frame = temp.read()  # Read the frame
      if not ret:
          break
      if fps_counter%fps==0:
        print(f"Current Step:{int(fps_counter/29)}")

        try:
            # Text Extraction
            original_stdout = sys.stdout            # Save the original stdout
            sys.stdout = open(os.devnull, 'w') # Redirect stdout to null device
            result = ocr.ocr(frame, cls=True)
            sys.stdout = original_stdout # Restore stdout

            # Turn Text to String
            i=0
            txts=[]
            for line in result:
              for i in range( len(line)):
                txts.append(line[i][1][0]) # Flatten the nested list and convert each element to a string

            result_string = ' '.join(txts)# Join the elements into a single string

            # Extract Ex Temp specifically
            match = re.search(r'Ex\.Temp:0*(\d+\.\d+)', result_string)
            if match:
                temperature = match.group(1)
                Ex_Temp.append(float(temperature))
            else:
                print("Ex Temp value not found in the image.")
                Ex_Temp.append('None')

            # Example for finding and appending numbers after "Vol" to the Vol list
            match = re.search(r'Vol:0*(\d+\.\d+|\d+)', result_string)
            if match:
                Vol.append(float(match.group(1)))
            else:
                print("Vol value not found in the image.")
                Vol.append('None')

            # Example for finding and appending numbers after "Cur" to the Cur list
            match = re.search(r'Cur:(\d+\.\d+|\d+)', result_string) # (\d+\.\d+|\d+) will match either a decimal number (e.g., 3.14) or a whole number (e.g., 42).
            if match:
                Cur.append(float(match.group(1)))
            else:
                print("Cur value not found in the image.")
                Cur.append('None')

            # Example for finding and appending numbers after "Pwr" to the Vol list
            match = re.search(r'Pwr:(\d+\.\d+|\d+)', result_string)
            if match:
                Pwr.append(float(match.group(1)))
            else:
                print("Pwr value not found in the image.")
                Pwr.append('None')

            # Example for finding and appending numbers after "Cap" to the Vol list
            match = re.search(r'Cap:(\d+\.\d+|\d+)', result_string)
            if match:
                Cap.append(float(match.group(1)))
            else:
                print("Cap value not found in the image.")
                Cap.append('None')

            # Example for finding and appending numbers after "Ene" to the Vol list
            match = re.search(r'Ene:(\d+\.\d+|\d+)', result_string)
            if match:
                Ene.append(float(match.group(1)))
            else:
                print("Ene value not found in the image.")
                Ene.append('None')

            # Example for finding and appending numbers after "Timing" to the Vol list
            match = re.search(r'Timing:(\d+):(\d+):(\d+)', result_string)
            if match:
                Timing.append(match.group(1))
            else:
                print("Timing value not found in the image.")
                Timing.append('None')

        except Exception as e:
            print(f"An error occurred: {e}")
            Extracted.append("Extraction failed")

      fps_counter=fps_counter+1

  temp.release()  # Release the video capture object

  # Convert the Extracted list to a DataFrame
  data = {
  'Vol': Vol,
  'Cur': Cur,
  'Pwr': Pwr,
  'Ene': Ene,
  'Cap': Cap,
  'Ex_Temp': Ex_Temp,
  'Timing': Timing}

  df = pd.DataFrame(data)

  # Specify the path and file name for the Excel file
  excel_file_path = '/content/drive/MyDrive/Colab Notebooks/Computer Science Projects/6 File Input-Output/Videos/Battery_Text_Excel'+'/'+Videos_names[n]+'/'+Videos_names[n]+'.xlsx'

  # Save the DataFrame to an Excel file
  df.to_excel(excel_file_path, index=False)
  print(f"Finished:{Videos_names[n]}")
  n=n+1
```
